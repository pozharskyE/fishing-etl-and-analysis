```python
import pandas as pd
import requests
import json
import time
```

```python
# https://dev.vk.com/ru/api/getting-started
with open('secret_vk_app_service_token.txt', 'r') as f:
    service_token = f.read()


api_method = 'groups.getMembers'
group_id = 56433000
sort_method = 'id_asc'



members_array = []

# because limit of count is 1000  => need to make request until all data will be recieved 
offset = 0
session = requests.Session()
while True:
    url = f"https://api.vk.com/method/{api_method}?group_id={group_id}&sort={sort_method}&fields=counters&offset={offset}&access_token={service_token}&v=4.3"


    response = session.get(url).json()['response']

    members_array.extend(response['items'])

    print(f"package {offset}-{offset+1000} received")

    offset += 1000
    # if offset > response['count']:
    #     break
    if offset + 1 > 20000:
        break
    
    # vk api have limit for number of requests per second (3) so 0.33 and adding error 0.02 => 0.35
    # time.sleep(0.35)
```
